# Como dividir los datos en datos de Entrenamiento y datos de Testeo

```python
import pandas as pd
from pandas import read_csv
from numpy.random import randn
import matplotlib.pyplot as plt
```

```python
data = read_csv('datasets/customer-churn-model/Customer Churn Model.txt')
data
```

## Dividir los datos utilizando la Distribución Normal

```python
a = randn(len(data))
plt.hist(a)
```

```python
# Vamos a quedarnos con los valores que cumplan una cierta condición
check = (a<0.8) # Nos quedaremos aproximadamente con el 80% de los valores
plt.hist(check.astype(int)) # Para plotear una variable booleana se usa el .astype(int)
```

Podemos observar arriba que el 80% de los valores aproximadamente son verdaderos, y más o menos el 20% son falsos.

Vamos a dividir el dataset en base a esto

```python
training = data[check]
testing = data[~check]
print('Tamaño Training dataset: ' + str(len(training)))
print('Tamaño Testing dataset: ' + str(len(testing)))
```

## Dividir Datos con la librería SKLEARN

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data, test_size=0.2)
print('Tamaño Training dataset: ' + str(len(train)))
print('Tamaño Testing dataset: ' + str(len(test)))
```

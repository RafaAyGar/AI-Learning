# Generar Variables Categóricas Y Continuas de forma aleatoria

```python
from numpy.random import randn, uniform, choice
import numpy as np
import pandas as pd
```

```python
gender = ['Male', 'Female']
income = ['Poor', 'Middle', 'Rich']
```

```python
n_rows = 500

gender_data = []
income_data = []

for i in range(0,500):
    gender_data.append(choice(gender))
    income_data.append(choice(income))

height = 170 + 10*randn(n_rows) # La altura seguirá una dist. normal de media 170 cm
weight = 70 + 10*randn(n_rows)
age = 31 + 10*randn(n_rows)
age = age.astype(int) # Redondeamos la edad para que no sea un número decimal
```

```python
data = pd.DataFrame(
    {
        'Gender' : gender_data,
        'Income' : income_data,
        'Height' : height,
        'Weight' : weight,
        'Age'    : age
    }
)
data.head()
```

## Podemos agrupar los datos por variables categóricas

```python
grouped_gender = data.groupby('Gender')
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

## También podemos agrupar por más de una variable categórica

```python
double_group = data.groupby(['Gender', 'Income'])
for names, groups in double_group:
    print(names)
    print(groups)
```

# Operaciones y Métodossobre datos agrupados

```python
double_group.sum()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
double_income = double_group['Income']
double_income.describe()
```

### Con aggregate podemos aplicar una función a cada columna y mostrarla

```python
double_group.aggregate(
    {
        'Weight' : np.sum,
        'Age' : np.mean,
        'Height' : np.std
    }
)
```

### Con lambda podemos hacer cualquier tipo de operación sobre la columna

```python
double_group.aggregate(
    {
        'Age' : np.mean,
        'Height' : lambda h:(np.mean(h))/np.std(h)
    }
)
```

### Podemos aplicar un conjunto de operaciones a todas las columnas

```python
double_group.aggregate([np.sum, np.mean, np.std])
```

### O una expresión lambda a todas las columnas

```python
double_group.aggregate([lambda x: np.mean(x)/np.std(x)])
```

# Filtrado de Datos

```python
double_group.sum()
```

### Imagina que quiero quedarme sólo con los grupos cuyas edades superan los 2400

```python
double_group['Age'].filter(lambda x: x.sum() > 2400)
```

# Transformación de Variables

Vamos a ver un ejemplo de transformación en el que normalizamos el valor de todas las columnas...

```python
zscore = lambda x : (x - x.mean()) / x.std()
double_group.transform(zscore)
```

# Cómo sustituir valores faltantes por la media utilizando expresiones lambda

```python
fillna_mean = lambda x : x.fillna(x.mean())
double_group.transform(fillna_mean)
```
